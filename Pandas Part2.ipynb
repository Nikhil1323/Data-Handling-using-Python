```python
#to get the data of female who have paid more than average.
df[(df['Sex'] == 'female') & (df['Fare'] > df['Fare'].mean())]
```

```python
df[(df['Sex'] == 'female') | (df['Fare'] > df['Fare'].mean())]
```

```python
#to get the data of Males who are below 30 years and paid less than average and managed to survive and belong to class3
df[(df['Sex'] == 'male') & (df['Age'] < 30) & (df['Fare'] < df['Fare'].mean()) 
   & (df['Survived'] == 1) & (df['Pclass']==3)]
```

```python
#to get the name of the passenger who have paid the maximum Fare
df[df['Fare'] == max(df['Fare'])]['Name']
```
