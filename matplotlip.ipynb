```python
x = ['a' , 'b', 'c','d', 'e']
```

```python
y = np.random.rand(5)
```

```python
plt.bar(x,y)
#representing categorical vs numeric data
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("Bar Graph")
```

```python

plt.figure(figsize=(6,2))
plt.plot(x,y,'--g')
plt.xlabel("This is my X Data")
plt.ylabel("This is my Y Data")
plt.title("This is my Title")
plt.grid()
plt.show()
```

```python
x = np.random.rand(50)
y = np.random.rand(50)

colours = np.random.rand(50)
sizes = 800 *np.random.rand(50)
plt.scatter(x,y,c = 'b',s = sizes, alpha = 0.5, marker = '*')
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("Random Data")
```
